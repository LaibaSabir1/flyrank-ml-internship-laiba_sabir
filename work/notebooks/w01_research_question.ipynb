## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
from IPython.display import Markdown, display

display(Markdown("""
**The decision:** Out of roughly 30,000 content pages, which ones should a content
reviewer look at first when they only have time to check a limited number per week?

**Who acts, and how:** A content editor/reviewer takes the ranked list this project
produces and manually inspects the top-ranked pages, deciding whether to refresh,
expand, or simply monitor each one.

**Cost of a wrong call:** If the ranking is wrong, there are two possible costs —
either the reviewer wastes limited time on pages that didn't actually need attention
(a false positive), or a page that is genuinely declining gets missed and continues
losing traffic unnoticed (a false negative). Since reviewer time is the scarce
resource, both errors are costly in different ways.
"""))
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

total_pages = len(df)
declining_rate = (df["trend_direction"].str.lower() == "down").mean()
median_impressions = df["impressions_90d"].median()

print("Total pages:", total_pages)
print("Declining rate (trend_direction == 'down'):", round(declining_rate, 3))
print("Median impressions (90d):", median_impressions)
display(Markdown(f"""
**What this shows:** Over half of all pages in this dataset ({declining_rate:.1%}) are currently labeled as declining, so prioritizing which ones to review first genuinely matters — this isn't a rare-event problem. On top of that, the starter pipeline shows a
simple hand-written rule only gets **24%** of its top-50 picks right (Precision@50),while a trained model gets **~74%** — roughly a 3x improvement. That gap is strong
evidence that a learned ranking can meaningfully beat a fixed rule on this data,which is exactly why this lane is worth pursuing.
"""))
```
